```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Sample car data
data = {
    'Car': ['Toyota', 'Honda', 'BMW', 'Audi', 'Ford', 'Chevrolet', 'Tesla'],
    'Price': [25000, 22000, 35000, 40000, 18000, 28000, 45000],
    'Mileage': [30000, 25000, 15000, 10000, 50000, 40000, 12000],
    'Engine Size (L)': [1.8, 2.0, 2.5, 3.0, 2.3, 2.0, 0.9]
}

# Create a DataFrame
df = pd.DataFrame(data)

# 1. Bar Plot: Price of Cars
plt.figure(figsize=(10, 6))
sns.barplot(x='Car', y='Price', data=df, palette='coolwarm')
plt.title('Car Prices')
plt.xlabel('Car')
plt.ylabel('Price (in $)')
plt.xticks(rotation=45)
plt.show()

# 2. Scatter Plot: Price vs. Mileage
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Mileage', y='Price', data=df, color='blue', s=100, marker='o')
plt.title('Price vs. Mileage')
plt.xlabel('Mileage (in miles)')
plt.ylabel('Price (in $)')
plt.show()


```
